```python
from env import config
from sqlalchemy import create_engine, text
```

```python
DATABASE_URL = config("DATABSE_URL", default=None)
DATABASE_URL is not None
```

```python
engine = create_engine(str(DATABASE_URL))
```

```python
DB_NAME="housing_db"
with engine.connect() as connection:
    result = connection.execute(text(f"CREATE DATABASE IF NOT EXISTS {DB_NAME}"))
    print(result)
```

```python
DB_NAME2="ai_travel_prices"
with engine.connect() as connection:
    result = connection.execute(text(f"CREATE DATABASE IF NOT EXISTS {DB_NAME2}"))
    print(result)
```
