```python
from env import config
from sqlalchemy import create_engine, text
from getpass import getpass
```

```python
DATABASE_URL = config("DATABSE_URL", default=None)
DATABASE_URL is not None
```

```python
engine = create_engine(str(DATABASE_URL))
```

```python
new_username = "mindsdb"
new_password = getpass()
```

```python
dbs = ["housing_db", "ai_travel_prices"]
with engine.connect() as connection:
    create_user_sql = f"CREATE USER IF NOT EXISTS '{new_username}'@'%' IDENTIFIED BY '{new_password}';"
    connection.execute(text(create_user_sql))

    for db in dbs:
        db_sql_perms = f"GRANT ALL PRIVILEGES ON {db}.* TO '{new_username}'@'%';"
        connection.execute(text(db_sql_perms))
    db_sql_new_dbs =f"GRANT CREATE ON *.* TO '{new_username}'@'%';"
    connection.execute(text(db_sql_new_dbs))
    connection.execute(text("FLUSH PRIVILEGES;"))
```

```python
with engine.connect() as connection:
    result = connection.execute(text(f"SHOW DATABASES"))
    print(result.all())
```
